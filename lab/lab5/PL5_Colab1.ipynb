```python
# MTConnect adapter sample for ME597 Lab5

import random
import sys
import time
import datetime
from data_item import Event, Sample
from mtconnect_adapter import Adapter


class MTConnectAdapter(object):

    def __init__(self, host, port):
        # MTConnect adapter connection info
        self.host = host
        self.port = port
        self.adapter = Adapter((host, port))

        # For samples
        self.a1 = Sample('a1')
        self.adapter.add_data_item(self.a1)

        self.a2 = Sample('a2')
        self.adapter.add_data_item(self.a2)

        self.a3 = Sample('a3')
        self.adapter.add_data_item(self.a3)

        # For events
        self.event = Event('event')
        self.adapter.add_data_item(self.event)

        # MTConnect adapter availability
        self.avail = Event('avail')
        self.adapter.add_data_item(self.avail)

        # Start MTConnect
        self.adapter.start()
        self.adapter.begin_gather()
        self.avail.set_value("AVAILABLE")
        self.adapter.complete_gather()
        self.adapter_stream()

    def adapter_stream(self):
        while True:
            try:
                # Generate random values
                a1 = random.uniform(-1, 1)
                a2 = random.uniform(-1, 1)
                a3 = random.uniform(-1, 1)

                self.adapter.begin_gather()

                self.a1.set_value(str(a1))
                self.a2.set_value(str(a2))
                self.a3.set_value(str(a3))

                self.adapter.complete_gather()

                print("{} RANDOM VALUE a1={} mm/s^2".format(
                    datetime.datetime.now(), a1))
                print("{} RANDOM VALUE a2={} mm/s^2".format(
                    datetime.datetime.now(), a2))
                print("{} RANDOM VALUE a3={} mm/s^2".format(
                    datetime.datetime.now(), a3))

                print(datetime.datetime.now(),
                      "Adapter 1 data items gathering completed...\n")

                time.sleep(1)  # Sampling period: 1 second

            except KeyboardInterrupt:
                print("Stopping MTConnect Adapter 1...")
                self.adapter.stop()
                sys.exit()


## ====================== MAIN ======================
if __name__ == "__main__":
    print("Starting Adapter 1!")
    MTConnectAdapter('127.0.0.1', 7878)
```

```python
# MTConnect adapter sample for ME597 Lab5

import random
import sys
import time
import datetime
from data_item import Event, Sample
from mtconnect_adapter import Adapter


class MTConnectAdapter(object):

    def __init__(self, host, port):
        # MTConnect adapter connection info
        self.host = host
        self.port = port
        self.adapter = Adapter((host, port))

        # For samples
        self.t1 = Sample('t1')
        self.adapter.add_data_item(self.t1)

        self.h1 = Sample('h1')
        self.adapter.add_data_item(self.h1)

        # For events
        self.event = Event('event')
        self.adapter.add_data_item(self.event)

        # MTConnect adapter availability
        self.avail = Event('avail')
        self.adapter.add_data_item(self.avail)

        # Start MTConnect
        self.adapter.start()
        self.adapter.begin_gather()
        self.avail.set_value("AVAILABLE")
        self.adapter.complete_gather()
        self.adapter_stream()

    def adapter_stream(self):
        while True:
            try:
                # Generate random values
                t1 = random.uniform(20, 25)
                h1 = random.uniform(30, 70)

                self.adapter.begin_gather()

                self.t1.set_value(str(t1))
                self.h1.set_value(str(h1))

                self.adapter.complete_gather()

                print("{} RANDOM VALUE t1={} °C".format(
                    datetime.datetime.now(), t1))
                print("{} RANDOM VALUE h1={} %".format(
                    datetime.datetime.now(), h1))

                print(datetime.datetime.now(),
                      "Adapter 2 data items gathering completed...\n")

                time.sleep(3)  # Sampling period: 3 seconds

            except KeyboardInterrupt:
                print("Stopping MTConnect Adapter 2...")
                self.adapter.stop()
                sys.exit()


## ====================== MAIN ======================
if __name__ == "__main__":
    print("Starting Adapter 2!")
    MTConnectAdapter('127.0.0.1', 7879)
```
