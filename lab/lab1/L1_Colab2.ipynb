### Data plotting

Plot a chart that includes temperature (in &deg;C, &deg;F) in the code box below:

```python
import matplotlib.pyplot as plt

df['time'] = pd.to_datetime(df["Timestamp"])
time = df['time'].dt.hour*3600 + df['time'].dt.minute*60 + df['time'].dt.second
t1 = time - time[0]
temperature = df['Temperature(°C)']

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Temperature graph', fontsize=20)
ax.set_xlabel('seconds', fontsize=10)
ax.set_ylabel('Degree of Celsius', fontsize=10)
ax.plot(t1, temperature, '-o')
plt.show()
```

```python
#Insert your coding below this line

temperature_c = df['Temperature(°C)']
temperature_f = df['Temperature(°F)']

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_title('Temperature graph (°C and °F)', fontsize=20)
ax1.set_xlabel('seconds', fontsize=10)
ax1.set_ylabel('Degree of Celsius', fontsize=10, color='blue')
ax1.plot(t1, temperature_c, '-o', color='blue', label='Celsius')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()  # 같은 x축을 공유하는 보조 y축 생성
ax2.set_ylabel('Degree of Fahrenheit', fontsize=10, color='red')
ax2.plot(t1, temperature_f, '-o', color='red', label='Fahrenheit')
ax2.tick_params(axis='y', labelcolor='red')

fig.tight_layout()
plt.show()
```
