## 2.1 Python programming for data collection from power meter


The entire schematic of the data communication using Modbus protocol in case of multiple masters
(computers) and a single slave (power meter) is shown in Figure 12. In the lab, your computers (laptop and Raspberry Pi) are on the router network. The Modbus RTU to Modbus TCP converter (gateway) is also on the same network with a static IP address. The ethernet cable (RJ45) is used for connection between the router and the gateway. The data communication sequence to read data from power meter is as follows.

1.	The computer (Raspberry Pi or laptop) requests read (data) holding registers using Modbus TCP protocol via the router to the gateway.
2.	The gateway converts Modbus TCP packet frame to Modbus RTU data frame and sends it to the slave.
3.	The slave responds to the request and sends Modbus RTU data frame to the gateway.
4.	The gateway converts the received Modbus RTU data frame to Modbus TCP packet frame.
5.	The gateway sends the converted packet to a specific IP host which is the computer requested data.
6.	The computer receives Modbus TCP packet frame and interprets the packet to the data.

To do this, we will use Python and ‘pymodbus’ module. If you are interested in ‘pymodbus’ module in detail, please visit the website (https://pymodbus.readthedocs.io/en/latest/, available on Aug. 5, 2024).

```
from pymodbus.client.sync import ModbusTcpClient
from pymodbus.constants import Endian
from pymodbus.payload import BinaryPayloadDecoder
import datetime

host = "192.168.1.100"  # Online student with Cisco VPN: "10.165.67.146"
port = 502

client = ModbusTcpClient(host, port=port)

if not client.connect():
    raise RuntimeError(f"Connect failed: {host}:{port}")

now = datetime.datetime.now()

freq_read = client.read_holding_registers(1536, count=2, unit=1)

if freq_read.isError():
    client.close()
    raise RuntimeError(f"Modbus read error: {freq_read}")

freq_reg = freq_read.registers

decoder = BinaryPayloadDecoder.fromRegisters(
    freq_reg,
    byteorder=Endian.Big,
    wordorder=Endian.Big
)


freq_value = decoder.decode_32bit_float()

print(f"{now}: Frequency is {freq_value} Hz")

client.close()


```

````python
import csv
import datetime
import time
import matplotlib.pyplot as plt
import pandas as pd
from pymodbus.client import ModbusTcpClient # Changed from pymodbus.client.sync
# from pymodbus.constants import Endian # Deprecated in newer pymodbus
from pymodbus.payload import BinaryPayloadDecoder

# --- [1. 설정 및 환경 변수] ---
# 보내주신 주석과 호스트 IP를 반영했습니다. 상황에 맞게 주석을 해제하거나 수정하세요.
host = "192.168.1.100"  # 기본 로컬 IP
# host = "10.165.67.146"  # Online student with Cisco VPN
port = 502
unit_id = 1

SAMPLING_PERIOD = 1.0  # 1초 주기 (sampling period 1 second)
TOTAL_DURATION = 60  # 최소 1분 동안 수집 (at least 1 minute)
CSV_FILENAME = "power_meter_data.csv"

# ⚠️ 장비 매뉴얼에 맞는 연속된 레지스터 혹은 개별 주소 맵을 설정합니다.
# (주파수 1536번을 기준으로 각각 32비트 float[count=2]이므로 2씩 증가한다고 가정했습니다)
ADDRESS_MAP = {
    "Frequency": 1536,
    "Voltage 1": 1538,
    "Current 1": 1540,
    "Power factor": 1542,
    "True power": 1544,
}


# --- [2. 레지스터 변환 함수 생성 (Hint 반영)] ---
def registers_to_float(registers):
    """지정된 레지스터 목록(input argument)을 32비트 float 값(output)으로 변환합니다."""
    decoder = BinaryPayloadDecoder.fromRegisters(
        registers, byteorder="big", wordorder="big" # Changed Endian.Big to "big"
    )
    return decoder.decode_32bit_float()


# --- [3. 데이터 분석 및 시각화 함수 (마지막 요구사항 반영)] ---
def analyze_and_plot(csv_file):
    print("\n" + "=" * 50)
    print(" 데이터 수집 완료! 시계열 분석 및 그래프 생성을 시작합니다.")
    print("=" * 50)

    # 저장된 CSV 파일 로드
    df = pd.read_csv(csv_file)

    # 1) 시간 도메인 특징값 계산 (Time-domain features: 평균값 계산)
    print("\n[시간 도메인 특징값 (평균값)]")
    for col in df.columns[1:]:  # Time [sec] 열을 제외한 데이터 열
        mean_val = df[col].mean()
        print(f" - {col} 평균: {mean_val:.4f}")

    # 2) 데이터 플롯팅 (Plot data)
    plt.figure(figsize=(12, 8))

    # Table 3 규격의 4가지 주요 지표를 각각 서브플롯으로 시각화
    metrics = [
        ("Frequency [Hz]", 221),
        ("Voltage 1 [V]", 222),
        ("Current 1 [A]", 223),
        ("True power [W]", 224),
    ]

    for column, pos in metrics:
        plt.subplot(pos)
        # 소수점 타임스탬프에서 시작 타임스탬프를 빼서 경과 시간(0초부터 시작)으로 플롯팅
        elapsed_time = df["Time [sec]"] - df["Time [sec]"].iloc[0]
        plt.plot(elapsed_time, df[column], "b-", linewidth=1.5)
        plt.title(column)
        plt.xlabel("Elapsed Time [sec]")
        plt.ylabel("Value")
        plt.grid(True)

    plt.tight_layout()
    plot_filename = "power_meter_plot.png"
    plt.savefig(plot_filename)
    plt.show()  # 코랩 화면에 그래프 즉시 출력
    print(f"\n📈 시계열 그래프 이미지가 '{plot_filename}'으로 저장되었습니다.")


# --- [4. 메인 데이터 수집 루프] ---
def main():
    client = ModbusTcpClient(host, port=port)

    if not client.connect():
        raise RuntimeError(f"Connect failed: {host}:{port}")

    print(f"장비 연결 성공 ({host}:{port}). 데이터 수집을 시작합니다...")

    # Table 3와 완전히 일치하는 CSV 헤더 구성
    headers = [
        "Time [sec]",
        "Frequency [Hz]",
        "Voltage 1 [V]",
        "Current 1 [A]",
        "Power factor [-]",
        "True power [W]",
    ]

    with open(CSV_FILENAME, mode="w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(headers)

        start_time = time.time()
        end_time = start_time + TOTAL_DURATION

        try:
            # 'while' 루프와 'time' 패키지를 사용한 주 주기 제어
            while time.time() < end_time:
                loop_start_time = time.time()

                # Table 3 예시 포맷과 일치하도록 소수점 6자리 고정밀 타임스탬프 생성
                data_row = [f"{loop_start_time:.6f}"]
                error_occurred = False

                # 순서대로 5가지 데이터 항목 읽기 및 변환
                for key in [
                    "Frequency",
                    "Voltage 1",
                    "Current 1",
                    "Power factor",
                    "True power",
                ]:
                    addr = ADDRESS_MAP[key]
                    result = client.read_holding_registers(
                        addr, count=2, unit=unit_id
                    )

                    if result.isError():
                        print(
                            f"[{datetime.datetime.now()}] Modbus read error at {key} (Addr: {addr})"
                        )
                        error_occurred = True
                        break

                    # 힌트대로 작성한 함수를 통해 float 변환 후 저장 행에 추가
                    val = registers_to_float(result.registers)
                    data_row.append(val)

                # 에러가 없는 정상 데이터 세트만 파일에 실시간 기록
                if not error_occurred:
                    writer.writerow(data_row)
                    f.flush()  # 실시간 유실 방지 코랩 버퍼 비우기
                    print(
                        f"[{loop_start_time:.2f}s] Freq: {data_row[1]:.4f} Hz | Volt: {data_row[2]:.2f} V | Power: {data_row[5]:.2f} W"
                    )

                # 연산 수행 속도를 감안하여 정확히 1초 주기를 맞추기 위한 딜레이 계산
                elapsed_loop = time.time() - loop_start_time
                sleep_time = SAMPLING_PERIOD - elapsed_loop
                if sleep_time > 0:
                    time.sleep(sleep_time)

        except KeyboardInterrupt:
            print("\n 사용자가 실행을 중단했습니다.")

        finally:
            client.close()
            print(f"\n데이터 수집 완료! 데이터가 '{CSV_FILENAME}'에 기록되었습니다.")

    # 과제 지문의 연습 문제 조건 수행: 그래프 플롯팅 및 특징값 계산 호출
    try:
        analyze_and_plot(CSV_FILENAME)
    except Exception as e:
        print(f"\n시각화/분석 수행 중 오류 발생: {e}")


if __name__ == "__main__":
    main()
````
