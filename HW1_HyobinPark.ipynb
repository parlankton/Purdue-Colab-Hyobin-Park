# [HW 1-2] Data Processing & Plotting
Plot the change in mean value every 0.01 seconds for each sensor signal in 'SensorData' above.
* Refer to ML1_Code3

```python

# Import (additional) packages
# No additional packages are required.
# Feature extraction
WindowSize = 128 # 0.01 seconds at 12,800 samples/second
# Corrected np.zeros arguments for shape and SensorData.shape access
Mean_Values = np.zeros((int(SensorData.shape[0] / WindowSize), SensorData.shape[1]))

# Corrected SensorData.shape access and Mean_Values indexing
for sensor in range(SensorData.shape[1]):
  for time in range(Mean_Values.shape[0]):
    Mean_Values[time, sensor] = np.mean(SensorData.iloc[WindowSize*time:WindowSize*(time+1), sensor])

TimeArr = np.arange(3, 13) / 100

# Plot graphs
plt.figure(figsize=(10,5))
# Corrected plotting arguments (Mean_Values indexing, linestyle, marker, markersize, markerfacecolor, markeredgecolor)
plt.plot(TimeArr, Mean_Values[:,0], ls='-', c='r', marker='o', ms=5, mfc='r', mec='r') # Acceleration
plt.plot(TimeArr, Mean_Values[:,1], ls='-', c='g', marker='o', ms=5, mfc='g', mec='g') # Voltage
plt.plot(TimeArr, Mean_Values[:,2], ls='-', c='b', marker='o', ms=5, mfc='b', mec='b') # Current
plt.grid()
plt.xlabel('time(s)')
plt.ylabel('Mean value')
# Corrected legend list and loc string
plt.legend(['Acceleration', 'Voltage', 'Current'], loc='best', fontsize=10)
plt.show()
```
